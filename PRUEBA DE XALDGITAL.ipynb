```python
###a)	¿Cuál fue el país que registró el mayor número de despegues?
import pandas as pd
archivo = "airNuevo.csv"
df = pd.read_csv(archivo, header=0, sep=",")
#print(df.columns)
df.columns = df.columns.str.strip()

# Verificar las columnas nuevamente
#print(df.columns)
#print(df['origen'].head())

df_despegues = df.groupby("origen").count()
df_despegues

df_despegues = df_despegues.sort_values("origen", ascending=False)
df_despegues

despegues_mayor=df_despegues.idxmax()
despegues_mayor
```

```python
#inciso d Determinar qué día presentó la mayor cantidad de retrasos y cuál la menor 
import pandas as pd
df1 = pd.read_csv('fecha.csv')         
df2 = pd.read_csv('retrasos.csv')
df3 = pd.read_csv('vuelos.csv')      


fecha_retrasos_df = pd.merge(df1,df2, on='vuelo', how='inner')
joined = pd.merge(fecha_retrasos_df,df3, on='vuelo', how='inner')


retrasos_por_dia = joined.groupby('dia').agg(total_retrasos=('retraso', 'sum')).reset_index()


dia_mayor_retraso = retrasos_por_dia.loc[retrasos_por_dia['total_retrasos'].idxmax()]
dia_menor_retraso = retrasos_por_dia.loc[retrasos_por_dia['total_retrasos'].idxmin()]

print(f"El día con mayor cantidad de retrasos:{dia_mayor_retraso}")
print(f"El día con menor cantidad de retrasos: {dia_menor_retraso}")

```

```python
### EJERCICIO 3 
import pandas as pd
df_fecha = pd.read_csv('fecha.csv')        
df_retrasos = pd.read_csv('retrasos.csv')
df_vuelos = pd.read_csv('vuelos.csv')       

# 2. primero realizo los merge para los df de los archivos 
fecha_retrasos = pd.merge(df_fecha, df_retrasos, on='vuelo', how='inner')
joined_final = pd.merge(fecha_retrasos, df_vuelos, on='vuelo', how='inner')


result_df = joined_final.groupby('vuelo').agg(conteo_vuelos=('vuelo', 'count'),
                                              suma_retrasos=('retraso', 'sum')).reset_index()


result_df['retrasos_acumulados'] = result_df['suma_retrasos'].cumsum()
result_df

```
